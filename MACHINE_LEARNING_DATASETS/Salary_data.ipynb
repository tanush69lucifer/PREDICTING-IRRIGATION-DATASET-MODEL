```python
df = pd.read_csv('https://raw.githubusercontent.com/arib168/data/refs/heads/main/spam.tsv',sep='\t')
df
```

```python
df = pd.read_table('https://raw.githubusercontent.com/arib168/data/refs/heads/main/spam.tsv')
df
```

```python
df['label'].value_counts()
# highly imbalanced dataset
# accuracy cannot be the primary metric
```

```python
x = df['message'].values
y = df['label'].values
```

```python
x.ndim
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
x_train_vect = vect.fit_transform(x_train)
x_test_vect = vect.transform(x_test)
```

```python
x_train_vect.ndim
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
nb.fit(x_train_vect,y_train)
```

```python
y_pred = nb.predict(x_test_vect)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
in1 = ['hi there nice to meet you, want to win free tickets to the IPL cricket match?']
op =vect.transform(in1)
nb.predict(op)
```

```python
in1 = ['win free tickets']
op =vect.transform(in1)
nb.predict(op)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,classification_report
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Pipelining - Combining 2 modules in a single step
from sklearn.pipeline import make_pipeline
model3 = make_pipeline(CountVectorizer(),MultinomialNB())
model3.fit(x_train,y_train)
y_pred3 = model3.predict(x_test)
y_pred3
```
